# Ïù¥Ìï≠Î∂ÑÌè¨
Ïù¥Ìï≠Î∂ÑÌè¨(Binomial Distribution)Îäî ```Í≥†Ï†ïÎêú ÌöüÏàòÏùò ÎèÖÎ¶ΩÏ†ÅÏù∏ ÏãúÎèÑ(trial)```ÏóêÏÑú Í∞Å ÏãúÎèÑÎßàÎã§ ÎèôÏùºÌïú ÏÑ±Í≥µ ÌôïÎ•† ùëù Î•º Í∞ÄÏßà Îïå, ÏÑ±Í≥µ ÌöüÏàòÎ•º ÎÇòÌÉÄÎÇ¥Îäî Ïù¥ÏÇ∞ ÌôïÎ•†Î∂ÑÌè¨ÏûÖÎãàÎã§. ÏòàÎ•º Îì§Ïñ¥, ÎèôÏ†ÑÏùÑ 10Î≤à ÎçòÏ°åÏùÑ Îïå ÏïûÎ©¥Ïù¥ ÎÇòÏò§Îäî ÌöüÏàòÎ•º Î™®Îç∏ÎßÅÌïòÎäî Îç∞ Ïù¥Ìï≠Î∂ÑÌè¨Î•º ÏÇ¨Ïö©Ìï† Ïàò ÏûàÏäµÎãàÎã§.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
import koreanize_matplotlib
````

````python
binom.pmf(1, 10, 0.5) # 10Î≤à Ï§ë 1Î≤à ÏïûÎ©¥Ïù¥ ÎÇòÏò¨ ÌôïÎ•†.
````

````python
n = 10
p = 0.5

values = np.arange(n+1)
binom.pmf(values, n, p)

idx = 0
for i in binom.pmf(values, n, p):
    print(f"ÏïûÎ©¥Ïù¥ {idx}Î≤à ÎÇòÏò¨ ÌôïÎ•† : {i * 100:.2f}%")
    idx += 1
````

````python
ps = binom.pmf(values, n, p)

plt.stem(ps)
plt.grid()
plt.show()
````

````python
# Ïù¥Ìï≠Î∂ÑÌè¨ ÏãúÎÆ¨Î†àÏù¥ÏÖò
n = 10
p = 0.5
samples = 100000

box = np.random.binomial(n, p, samples)

plt.hist(box)
plt.show()
````

````python
# ÎàÑÏ†Å ÌôïÎ•† Î∂ÑÌè¨
n = 10
p = 0.5

values = np.arange(11)
ps = binom.cdf(values, n, p)

idx = 0
for i in ps:
    print(f"ÏïûÎ©¥Ïù¥ {idx}Î≤à Ïù¥ÌïòÎ°ú ÎÇòÏò¨ ÌôïÎ•† : {i*100}%")
    idx += 1
````

````python
plt.step(values, ps)
plt.show()
````
